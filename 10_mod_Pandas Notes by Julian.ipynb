Seleccion de datos con condicion

```
filas='ventas1 ventas2 ventas3'.split() #otra foema de crear los valores
columnas='zonaA zonaB zonaC'.split()
datos=[[120,340,250],[210,450,250],[310,210,140]]
```

```
#creamos el dataframe
dataframe=pd.DataFrame(datos,filas,columnas)
dataframe
```

```
#condicion
condicion=dataframe>200
#utilizamos la condicion para extraer datos
dataframe[condicion]
```

```
#por condicion dentro de una columna
condicion=dataframe['zonaA'] > 200
dataframe[condicion]
```

```
# creando condicion dobles
condicion= (dataframe['zonaA'] > 200) & (dataframe['zonaB'] > 300)
dataframe[condicion]
```

```
#ahora aplicamos la condicion anterios pero en una columna determinada
dataframe[condicion]['zonaB']
```

```
#en varias columnas
dataframe[condicion][['zonaB','zonaC']]
```

```
dataframe
```

```
#cambiamos los nombres de las filas
nuevas_filas='dia1 dia2 dia3'.split()
dataframe['dias']=nuevas_filas
```

```
dataframe
```

```
#ahora se pasa esa columa al index para cabiar el nombre de filas
dataframe.set_index('dias')
```

```
#para ejecutarlo de forma permanente
dataframe=dataframe.set_index('dias')
dataframe
```

```
#creamos un dataframe a partir de un diccionario
diccionario={'A':[4,5,np.nan], 'B': [6,1,5], 'C': [np.nan,4,np.nan]}
dataframe=pd.DataFrame(diccionario)
dataframe
```

```
#Para borrar los nulos    OJO---- borra cualquier fila donde este nulo
dataframe.dropna()
```

```
dataframe.dropna(axis=1)
#con inplace lo ejecuta definitivamente
```

```
#para rellenar los valores nulos con algun valor
dataframe.fillna(value=100)
```

```
#o rellenarlo con el valor medio
valor_medio=dataframe.mean()
dataframe.fillna(value=valor_medio)
```
