```
#Import the libraries

import pandas as pd
import numpy as np
```

```
#Import the dataset using a URL
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
chipo_data = pd.read_csv(url, sep = '\t')
chipo_data.head(5)  #Show the first 5 rows 
```

Count the number of rows and columns of a DataFrame

```
#One solution

total_rows= chipo_data.shape[0]
total_cols= chipo_data.shape[1]
print("Number of Rows: "+str(total_rows))
print("Number of Columns: "+str(total_cols))
```

```
#Other alternative that I found

total_rows=len(chipo_data.axes[0])
total_cols=len(chipo_data.axes[1])
print("Number of Rows: "+str(total_rows))
print("Number of Columns: "+str(total_cols))
```

Show the columns of the dataframe

```
chipo_data.columns
```

To show the range of Index

```
chipo_data.index
```

Using Group by

```
#The most-ordered item

c = chipo_data.groupby('item_name').sum()
c = c.sort_values(['quantity'], ascending=False) 
c.head(1)
```

```
#The most-ordered item
# Other way to do that

c = chipo_data.groupby('item_name')['quantity'].sum()
c = c.sort_values(ascending=False)
c.head(1)
```

```
#The most ordered item in the choice_description column

c = chipo_data.groupby('choice_description').sum()
c = c.sort_values(['quantity'], ascending=False) 
c.head(1)
```

```
#How many items were orderd in total?
total_items_orders = chipo_data.quantity.sum()
total_items_orders
```

```
#How many orders were made in the period?
orders = chipo_data.order_id.value_counts().count()
orders
```

###Filtering and Sorting Data

```
#Import the dataset using a URL
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
chipo = pd.read_csv(url, sep = '\t')
chipo.head(5)  #Show the first 5 rows 
```

```
#Sort by the name of the item
chipo.sort_values(by = "item_name")
```

```
#What was the quantity of the most expensive item ordered?
chipo.sort_values(by = "item_price", ascending = False).head(1)

```

```
#How many times was a Veggie Salad Bowl ordered?
chipo_salad = chipo[chipo.item_name == "Veggie Salad Bowl"]
len(chipo_salad)
```
