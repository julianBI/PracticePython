```
# List Slicing to Reverse an Array in Python
import numpy as np

def arrays(arr):
    # complete this function
    # use numpy.array
    arr = np.array(arr[ : : -1], dtype = "float")
    return(arr)

arr = input().strip().split(' ')
result = arrays(arr)
print(result)
```

```
# Another way to reverse an Array (Using Flip method)
import numpy as np

def arrays(arr):
    # complete this function
    # use numpy.array
    arr = np.array(arr, dtype = "float")
    arr = np.flip(arr)  #To reverse the array
    return(arr)

arr = input().strip().split(' ')
result = arrays(arr)
print(result)
```

```
# Another way to reverse an Array (Using Flipud method)
import numpy as np

def arrays(arr):
    # complete this function
    # use numpy.array
    new_arr=np.array(arr)
    arr = np.flipud(new_arr)  #To reverse the array
    return(arr)

arr = ("J U L I A N").strip().split(' ') # string variable
result = arrays(arr)
print(result)
```
