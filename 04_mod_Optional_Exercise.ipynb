###1.Write a Python program to reverse a string.

```
#This is an example using extended slice syntax

message='This is a message'
print(message[::-1])
```

```
#Using a sorted() function. It returns a sorted list of the specified iterable object
message=['This', 'is', 'a' , 'message']
newmessage=sorted(message, reverse=True)
print (message)
print (newmessage)
```

###I found other way to recreate the same result

```
message = "Python" # initial string
reversedString=[]
index = len(message) # calculate length of string and save in index
while index > 0: 
    reversedString += message[ index - 1 ] # save the value of str[index-1] in reverseString
    index = index - 1 # decrement index
print(reversedString) # reversed string
```

###2.Write a Python function to multiply all the numbers in a list.

```
#Write a Python code to multiply all the numbers in a list.

lista=[1,2,3,4]
resul=1
contador=0
for con in lista:
  resul=resul * con
  contador=contador+resul
print (resul)
```

###2.1 I did a function and ask for the items of the list

```
# In this part I've created a function
def operation (lista):
  resul=1
  contador=0
  for con in lista:
    resul=resul * con
    contador=contador+resul
  return resul

# In this part I'm going to ask for the list
lst = []
n = int(input("Enter number of elements : "))
for i in range(0, n):
  element = int(input("Element number  %d  : " % i ))
  lst.append(element) # adding the element

# In this part call the function
result= operation(lst)
print("This is the reslt of multiply all the numbers in a list ---> %d" %result)
```

###Write a Python function that accepts a string and calculate the number of upper case letters and lower

```
# In this section I set the variables
message='The Power To Know'
message_strip=message.replace(" ", "") #the idea here is to rid off all blank spaces

contadorm=0
contadorn=0

# This loop let me know if the leter is lower or upper case
for i in message_strip:
  if i.upper()==i:
    contadorm=contadorm+1
  if i.lower()==i:
    contadorn=contadorn+1

# This code let me know how much blank spaces have the original variable
ocurrencias = message.count(" ")       

print('Total Uppercase letters = %d' % contadorm)
print('Total Lowercase letters = %d' % contadorn)
print('Total of blank spaces = %d' % ocurrencias)
```
