#Examples with itertools

```
# Impor the library
from itertools import product 
from itertools import permutations
```

```
# to get a list
arr1 = [1, 2, 3] 
arr2 = [5, 6, 7]
```

```
# print out the product
print (list(product(arr1, arr2)))
```

####Do the same exercise but using Input function to get values

```
#To the same bu using input function to get the list
  
def productf(A,B):
  # To get an integer value from  (A)  Using list comprehension
  A = [int(i) for i in A] 
  # To get an integer value from (B) Using map function
  B = list(map(int, B)) 
  return list(product(A,B))

# get the values
A = list(input("Input numbers follow by (,): ").split(','))
B = list(input("Input numbers follow by (,): ").split(','))

#print out the result
print(productf(A,B))

```
