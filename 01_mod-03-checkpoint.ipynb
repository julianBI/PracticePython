```
#Make a list of the first ten multiples of ten (10, 20, 30... 90, 100) and create a list with them
list100=[]
for i in range(1,11):
  list100.append(i*10)
print('This is a list using normal way \n')
print (list100)

#use comprehensions statement to optimize the code before
list100=[i*10 for i in range(1,11)]
print('\nThis is a list using comprenhensions \n')
print (list100)
```
