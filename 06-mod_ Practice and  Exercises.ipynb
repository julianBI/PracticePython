###Regular expression
Basic Example

```
import re #import regular expression module

cadena='can,dan,man,ran,pan,fan'

#The idea is to match only three words: can,man,fan

#define the rule:
patron='[a-z]*[^drp]an'
re.findall(patron,cadena)
```

```
#only for testing I cahnge the order of the patron.
#As a result of this I got the same result

patron='[^drp]an*[a-z]'
re.findall(patron,cadena)
```

####Example2
Create a variable with date information.
The idea is to get the full date and only the year using Nested Group.

```
#create a variables
dates='Jan 1982, Sep 1972, Nov 2011'
#define the rule:
patron='(\w+ (\d+))'
re.findall(patron,dates)
```

#### Valid email addres

This example only work using this pattern
- The username starts with an English alphabetical character, and any subsequent characters consist of one or more of the following: alphanumeric characters, -,., and _.
- The domain and extension contain only English alphabetical characters.
- The extension is 3 characters in length.


```
# This variable contain the email
email='u.rAa.me@donnnnain.col1'

#define the rule:
patron='^[A-Za-z \_ \- \.]+@[A-Za-z]+[\.]+[A-Za-z]{3}$'
match=re.findall(patron,email)


if not match:
  print('This e-mail is not valid ...... sorry ;-)')
else:
  print('This e-mail is valid : {} '.format (match))


```
